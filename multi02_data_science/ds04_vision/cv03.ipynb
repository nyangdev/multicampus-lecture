```python
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/roselia.png")

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
# b
cv.imshow("blue", img[:, :, 0])

cv.waitKey()
cv.destroyAllWindows()
```

```python
cv.imshow("green", img[:, :, 1])

cv.waitKey()
cv.destroyAllWindows()
```

```python
cv.imshow("red", img[:, :, 2])

cv.waitKey()
cv.destroyAllWindows()
```

```python
plt.imshow(img[:, :, 0])
```

```python
plt.imshow(img_plot[:, :, 0])
```

```python
img_blue = img[:, :, 0]
img_green = img[:, :, 1]
img_red = img[:, :, 2]
```

```python
img_bgr = cv.vconcat([img_blue, img_green, img_red])

cv.imshow("bgr_v", img_bgr)

cv.waitKey()
cv.destroyAllWindows()
```

```python
img_bgr = cv.hconcat([img_blue, img_green, img_red])

cv.imshow("bgr_v", img_bgr)

cv.waitKey()
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(15, 5))

blue = cv.calcHist([img], [0], None, [256], [0, 256])
green = cv.calcHist([img], [1], None, [256], [0, 256])
red = cv.calcHist([img], [2], None, [256], [0, 256])

plt.subplot(131)
plt.plot(blue, color="b")

plt.subplot(132)
plt.plot(blue, color="g")

plt.subplot(133)
plt.plot(blue, color="r")
```
