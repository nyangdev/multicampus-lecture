```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("../images/headache.png")

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
#  
# cv.threshold( ,   ,   ,  )

th, img_th = cv.threshold(img, 100, 255, cv.THRESH_BINARY)
```

```python
cv.imshow("threshold", img_th)

cv.waitKey()
cv.destroyAllWindows()
```

```python
#  (OTSU) :   
#  
th, img_th = cv.threshold(img[:, :, 2], 0, 255, cv.THRESH_OTSU)
print(f"    {th}")
```

```python
cv.imshow("threshold-otsu", img_th)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# gamma 
def gamma(image, gamma=1.0):
    image_gm = image / 255.0
    return np.uint8(255 * (image_gm ** gamma))
```

```python
gamma_list = np.hstack((gamma(img, 0.5), gamma(img, 0.75),
                        gamma(img), gamma(img, 2.0), gamma(img, 3.0)))

cv.imshow("gammas", gamma_list)

cv.waitKey()
cv.destroyAllWindows()
```
