```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/pika.png")
```

```python
# dsize : 절대적인 크기 변경
# fx, fy : 상대적인 크기 변경
img_half = cv.resize(img, dsize=(0, 0), fx=0.5, fy=0.5)
```

```python
cv.imshow("half", img_half)

cv.waitKey()
cv.destroyAllWindows()
```

```python
img_500 = cv.resize(img, dsize=(500, 500))
```

```python
cv.imshow("500", img_500)

cv.waitKey()
cv.destroyAllWindows()
```

```python
print(img.shape)
print(img_half.shape)
print(img_500.shape)
```

```python
img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
img_center = img[
    img.shape[0]//4 : 3 * img.shape[0]//4,
    img.shape[0]//4 : 3 * img.shape[1]//4,
    :
]
```

```python
cv.imshow("center", img_center)

cv.waitKey()
cv.destroyAllWindows()
```

```python
cv.vconcat([img, img_center])
```

```python
img_center_resize = cv.resize(img_center, (img.shape[0], img.shape[1]))
img_h = cv.hconcat([img, img_center_resize])
```

```python
cv.imshow("img_h", img_h)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# 바둑판 형태
img_list = [
    [img, img, img],
    [img, img_center_resize, img],
    [img, img, img]
]
```

```python
img_list_concat = cv.vconcat([cv.hconcat(img_list_h) for img_list_h in img_list])
```

```python
cv.imshow("concat", img_list_concat)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# 회전
img_90 = cv.rotate(img_plot, cv.ROTATE_90_CLOCKWISE)
```

```python
plt.imshow(img_90)
```

```python
img_90_re = cv.rotate(img_plot, cv.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(img_90_re)
```

```python
# 반전
img_flip = cv.flip(img_plot, 0)
plt.imshow(img_flip)
```

```python
img_flip = cv.flip(img_plot, 1)
plt.imshow(img_flip)
```

```python
img_flip = cv.flip(img_plot, -1)
plt.imshow(img_flip)
```
