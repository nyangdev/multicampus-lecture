```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/logo.png")

img = cv.resize(img, dsize=(0, 0), fx=3, fy=3)

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
# 
kernel = np.ones((5, 5), np.uint8)
erode = cv.erode(img, kernel, iterations=2)

cv.imshow("erode", erode)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# 
kernel = np.ones((5, 5), np.uint8)
dilates = cv.dilate(img, kernel, iterations=2)

cv.imshow("dilates", dilates)

cv.waitKey()
cv.destroyAllWindows()
```

```python
"""
cv.MORPH_CLOSE      : dilation()  Erosion() 
cv.MORPH_OPEN       : Erosion  Dilation 
cv.MORPH_BLACKHAT   : Closing    
cv.MORPH_GRADIENT   : Dilation Erosion   
cv.MORPH_TOPHAT     : Opening    
"""

kernel = np.ones((5, 5), np.uint8)
ref, img_th = cv.threshold(img, 110, 255, cv.THRESH_BINARY)
gradient = cv.morphologyEx(img_th, cv.MORPH_GRADIENT, kernel)

result = cv.vconcat([img, gradient])

cv.imshow("gradient", result)

cv.waitKey()
cv.destroyAllWindows()
```
