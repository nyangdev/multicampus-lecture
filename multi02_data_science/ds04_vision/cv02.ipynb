```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/headache.png")
```

```python
plt.imshow(img)
```

```python
#  :  /  :  /  : 
img.shape
```

```python
# BGR (opencv) <-> RGB (matplotlib)
b, g, r = cv.split(img)

#  RGB   
img_rgb = cv.merge([r, g, b])

plt.imshow(img_rgb)
```

```python
img_rgb2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb2)
```
