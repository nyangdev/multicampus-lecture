```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
x
```

```python
def binary_step(x):
    return np.array(x >= 0)
```

```python
plt.plot(x, binary_step(x))
```

```python
def gaussian(x, mu=0, sigma=1, a=1):
    return a * np.exp(-(x-mu)**2 / sigma**2)
```

```python
plt.plot(x, gaussian(x))
plt.plot(x, gaussian(x, 2, 3, 0.5))
```

```python
# 이진분류의 output layer : 0 ~ 1 사이의 값 (확률)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, sigmoid(x))
```

```python
# 다중 분류 : 모두 더하면 1 (확률)
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.array([0.3, 2.9, 4.0])
y = softmax(x)

print(y)
print(y.sum())
```

```python
# sigmoid 이진분류
# softmax는 다중분류
# 각 개념 꼭 암기해줘야함
```

```python
# hidden layer에서 주로 사용
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5, 5, 0.1)
plt.plot(x, relu(x))
```
