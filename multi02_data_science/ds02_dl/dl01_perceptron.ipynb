perceptron : neuron을 본따서 만든 객체.

- 여러 개의 입력을 받아서 하나로 출력

    ex) Y = W1 * X1 + W2 * X2 + b (입력이 2개)

W : 가중치 (weight)

b : 편향 (bias)

```python
import numpy as np
```

```python
def perceptron(x, w, b):
    result = np.sum(w * x) + b

    # binary step
    if result < 0:
        return 0
    else:
        return 1
```

```python
x_list = np.array([[0,0], [0,1], [1,0], [1,1]])
```

```python
# and
# and 값을 얻기위해서 w랑 b 값을 준 것
w = np.array([0.5, 0.5])
b = -0.7

for x in x_list:
    print(f"{x} : {perceptron(x, w, b)}")
```

```python
# or
w = np.array([0.5, 0.5])
b = -0.4
for x in x_list:
    print(f"{x} : {perceptron(x, w, b)}")
```

```python
# nand : not + and = 1, 1, 1, 0
# and 값 거꾸로
w = np.array([-0.5, -0.5])
b = 0.7

for x in x_list:
    print(f"{x} : {perceptron(x, w, b)}")
```

XOR 위키백과 그림

![](https://upload.wikimedia.org/wikipedia/commons/a/a2/254px_3gate_XOR.jpg)

```python
# xor
for x in x_list:
    nand_res = perceptron(x, [-0.5, -0.5], 0.7)
    or_res = perceptron(x, [0.5, 0.5], -0.4)
    xor_res = perceptron(np.array([nand_res, or_res]), [0.5, 0.5], -0.7)
    print(f"{x} : {xor_res}")
```
