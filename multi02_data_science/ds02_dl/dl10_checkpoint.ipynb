```python
import os
from datetime import datetime

from tensorflow.keras.datasets.mnist import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
x_train_scale = x_train / 255.0
x_test_scale = x_test / 255.0
```

```python
model = Sequential([
    Flatten(),
    Dense(8, activation="relu"),
    Dense(32, activation="relu"),
    Dense(10, activation="softmax")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="adam",
              metrics=["acc"])
```

```python
save_dir = "dl10_checkpoins"

if os.path.exists(save_dir):
    pass
else:
    os.mkdir(save_dir)
```

```python
now = datetime.now()
save_time = f"/{now.year}_{now.month}_{now.day}_{now.hour}_{now.minute}"
save_time
```

```python
# .h5 (옛날 버전) -> .keras (현재)
checkpoint = ModelCheckpoint(filepath=save_dir + save_time + "_{epoch}_{val_loss:.4f}.keras",
                             monitor="val_loss",
                             save_best_only=True)
```

```python
result = model.fit(x_train_scale, y_train, epochs=100,
                   batch_size=100, callbacks=[checkpoint],
                   validation_split=0.2)
```

```python
model.evaluate(x_test_scale, y_test)
```
