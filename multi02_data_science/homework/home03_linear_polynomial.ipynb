```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# csv 불러오기
df = pd.read_csv("../datasets/weight_height.csv", encoding="euc-kr")
df
```

```python
df = df[["학교명", "학년", "성별", "키", "몸무게"]]
df
```

```python
df.isnull().sum()
```

```python
# df["학교명"] -> 초등학교=0, 중학교=6, 고등학교=9 + df["학년"]
df["grade"] = df["학교명"].map(lambda x : 0 if x.endswith("초등학교")
                            else 6 if x.endswith("중학교")
                            else 9) + df["학년"]
```

```python
df
```

```python
df.drop(["학교명", "학년"], axis="columns", inplace=True)
```

```python
df.columns = ["gender", "height", "weight", "grade"]
df
```

## 남학생의 몸무게를 가지고 키를 예측하자

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
print(df.isna().sum())
```

```python
plt.scatter(df["weight"], df["height"], marker="o")
```

```python
gender_condition = df["gender"] == "남"

x = df[gender_condition][["weight"]]
y = df[gender_condition][["height"]]
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
predict
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[100]])
```

```python
# 숙제: polynomial을 사용해서 직선이 아니라 곡선 데이터로 나오도록 만들기
```

```python
plt.scatter(x_test, y_test, color="r")
# x값 : x_test 데이터
# y값 : x_test 데이터에 대한 model의 예측값
plt.plot(x_test, model.predict(x_test), label="linear regression")
plt.legend()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 1학년
grade1 = df[df["grade"] == 1]
```

```python
print(grade1)
```

```python
x = grade1[["weight"]]
y = grade1[["height"]]
```

```python
x
```

```python
y
```

```python
# 1학년 학생들의 몸무게를 가지고 키를 예측한 곡선 그래프 그리기
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
poly = PolynomialFeatures(degree=3)
poly.fit(x_train)
```

```python
x_train_poly = poly.transform(x_train)
```

```python
model = LinearRegression().fit(x_train_poly, y_train)
```

```python
import numpy as np
```

```python
x_range = np.linspace(x["weight"].min(), x["weight"].max(), 100).reshape(-1, 1)
```

```python
x_range.shape
```

```python
predict = model.predict(poly.transform(x_range))
```

```python
model.score(poly.transform(x_test), y_test)
```

```python
plt.scatter(x_train, y_train, color="pink")
plt.plot(x_range, model.predict(poly.transform(x_range)))
plt.legend()
plt.show()
```
