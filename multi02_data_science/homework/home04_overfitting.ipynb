```python
import pandas as pd
import numpy as np
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
row_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)
data = np.hstack([row_df.values[::2, :], row_df.values[1::2, :2]])
target = row_df.values[1::2, 2]
```

```python
x = pd.DataFrame(data, columns=["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"])

y = pd.DataFrame(target, columns=["PRICE"])
```

LinearRegression, Ridge, Lasso, ElasticNet 학습하여 비교하기

score, coef_, intercept_

```python
# 숙제 : LinearRegression, Ridge, Lasso, ElasticNet 학습 및 비교 -> score, coef_, intercept_
```

```python
print(x)
print(y)
```

```python
# LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=1)
```

```python
model.fit(x_train, y_train)
```

```python
# 모델이 데이터를 얼마나 잘 예측하는지
model.score(x_test, y_test)
```

```python
# 각 변수가 주택가격에 미치는 영향력
model.coef_
```

```python
# 모든 변수가 0일때 기본 가격 (y절편)
model.intercept_
```

```python
# Ridge
```

```python
from sklearn.linear_model import Ridge
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=1)
```

```python
# 규제 강도 조절
model = Ridge(alpha=0.5)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# 규제강도 재조절
model = Ridge(alpha=0.8)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Lasso
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
from sklearn.linear_model import Lasso
```

```python
model = Lasso(alpha=0.5)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# ElasticNet
# Ridge와 Lasso를 절반씩 섞은 회귀 모델
```

```python
# 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet(alpha=0.5, l1_ratio=0.5)
# l1_ratio에서 0은 Ridge, 1은 Lasso
# 0.5는 각 규제를 반반씩 섞음
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```
