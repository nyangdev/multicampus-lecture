tf-idf : Term Frequency - Inverse Term Frequency

tf : 단어 빈도 -> 문서를 잘 설명하는 척도

idf : 역 문서 빈도 -> 전체 말뭉치 (corpus) 에서 얼마나 희귀한지 (자주 사용될 수록 값이 낮아짐)

단어에 가중치를 부여 (일반적인 단어 / 중요한 단어 비교)

- 자주 사용될수록 (많은 문서에서 등장할수록) 낮은 값 : 일반적인 단어

- 희귀하게 사용될수록 (특정 문서에서만 등장할수록) 높은 값 : 중요한 단어

```python
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corpus = ["A hundred bad days made a hundred good stories",
            "A hundred good stories make me interesting at parties"]
```

```python
cv = CountVectorizer(stop_words=stopwords.words("english"))
cv.fit(corpus)

bow = cv.transform(corpus)
df = pd.DataFrame(bow.toarray(), index=corpus, columns=cv.get_feature_names_out())
df
```

```python
tfv = TfidfVectorizer(stop_words=stopwords.words("english"))
tfv.fit(corpus)
```

```python
tf_idf = tfv.transform(corpus)
```

```python
tf_idf.toarray()
```

```python
# tf * idf 한 값 (= 해당 말뭉치를 가장 잘 표현하는 값)
# == 해당 단어의 가중치를 만든 값
tf_idf.toarray()
```

```python
tfv.get_feature_names_out()
```

```python
tfv.vocabulary_
```
