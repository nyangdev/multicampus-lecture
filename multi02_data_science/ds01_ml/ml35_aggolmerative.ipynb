agglomerative : 병합 군집 - 계층 기반 클러스터링

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
```

```python
x, y = make_blobs(random_state=1)
```

```python
predict = AgglomerativeClustering(n_clusters=3).fit_predict(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=predict)
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
x, y = make_blobs(random_state=0, n_samples=10)

linkage_array = ward(x)
linkage_array
```

```python
dendrogram(linkage_array)

ax = plt.gca()
bounds = ax.get_xbound()

ax.plot(bounds, [7.5, 7.5], "--", c="k")
ax.plot(bounds, [5.5, 5.5], "--", c="k")

ax.text(bounds[1], 7.5, "n_cluster: 2", va="center")
ax.text(bounds[1], 5.5, "n_cluster: 3", va="center")
```

병합 방법
- ward linkage : 전체 클러스터의 분산이 가장 작게 증가하도록 두 클러스터를 선택
- single linkage : 가장 가까운 두개의 데이터 사이의 거리를 이용하여 두 클러스터를 선택
- complete linkage : 가장 먼 두개의 데이터 사이의 거리를 이용하여 두 클러스터를 선택
- average linkage : 모든 데이터의 평균거리를 사용하여 두 클러스터를 선택
