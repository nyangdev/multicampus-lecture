lightGBM (Light Gradient Boosting Machine) : MSR (MicroSoft Research) 팀에서 발표

xgboost와 비슷 (더 빠르고 더 가볍다)

-> XGBoost 등의 결정트리 관련 경사 부스팅은 매우 높은 정밀도를 기대하는 한편, 결정 나무를 만들 때의 "분할"에 시간이 걸리는 것이 단점.

lightGBM 특징 (각 개념들 숙제!!)

- leaf wise : 학습 속도 향상, 복잡도 감소
- histogram based : 연속값의 특징량(feature)을 히스토그램화 하여 bin 단위로 분할 -> 학습 속도 향상
- Gradient based one-side sampling (GOSS) : 기울기가 큰 데이터는 모두 사용, 기울기가 작은 데이터는 랜덤 샘플링 -> 학습속도 향상
- Exclusive Freture Bundling (EFB) : 복수의 feature를 bundle하고 정리해 하나의 feature와 같이 취급 -> 복잡도 감소

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

import numpy as np
```

```python
cancer = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=1)
```

```python
"""
boosting_type : default "gbdt"
    gbdt : gradient boosting decision tree
    dart : dropouts meet multiple addicitive regression trees
    rf : random forest
"""

model = lgb.LGBMClassifier(boosting_type="gbdt", max_depth=5, n_jobs=-1, random_state=1)
```

```python
eval_set = [(x_test, y_test)]
model.fit(x_train, y_train, eval_set=eval_set)
```

```python
predict = model.predict(x_test)
predict
```

```python
model.score(x_test, y_test)
```
