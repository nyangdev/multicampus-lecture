나이브 베이즈

- 사전확률 및 추가정보를 기반으로 사후 확률을 추론 -> 베이즈 추정 기반 분류

- 종속변수 각 범주의 등장 빈도를 사전확률(prior)로 설정이 중요

- 각 데이터의 사전 확률(prior)을 기반으로 사후 확률(posterior)을 계산

1. 가우시안 (GaussianNB): 연속적인 특성 (정규분포)
2. 베루누이 분포 (BernoulliNB) : 이진 특성
3. 다항 분포 (MultinormialNB) : 카운트 특성 (ex. 단어의 갯수)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
accuracy_score(y_test, predict)
```

```python
# 예측 확률값
proba = model.predict_proba(x)
proba
```
