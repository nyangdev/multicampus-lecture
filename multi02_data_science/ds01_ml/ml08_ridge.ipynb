```python
import pandas as pd
import numpy as np
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
row_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)
data = np.hstack([row_df.values[::2, :], row_df.values[1::2, :2]])
target = row_df.values[1::2, 2]
```

```python
x = pd.DataFrame(data, columns=["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"])

y = pd.DataFrame(target, columns=["PRICE"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, Y_test = train_test_split(x, y, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import Ridge
```

```python
# alpha : 값이 커질수록 규제 강도가 높아진다 (과적합 완화)
model = Ridge(alpha=0.5)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, Y_test)
```

```python
# weight
print(model.coef_)

# bias
print(model.intercept_)
```

정규화 (Regulaarization) 선형회귀

- 가중치 (W) 에 대한 제약 조건을 추가함으로 모델이 과적합되는 것을 방지

- 과적합이 발생한다는 말은, 모델의 가중치도 과하게 증가한다는 말 => 가중치의 크기를 제한하여 해결

- 과적합 : 학습 데이터는 거의 완벽하게 예측하지만, 새로운 데이터의 예측은 안정적이지 않다.

- 예측값들이 실제 타겟 값과 전반적으로 멀리 떨어져 있으면 편향(bias) 높다.

- 예측값들이 전반적으로 서로 멀리 흩어져 있으면 결과의 분산(variance)이 높다.

- trade-off : 편향과 분산이, 한쪽이 증가하면 다른쪽은 감소 or 한쪽이 증가하면 다른 한쪽도 증가
    - 모델학습을 처음 할 때 모델이 정교해지면(복잡도) 편향이 감소해서 모델의 전체 오차가 감소
    - 단, 어느 시점부터는 모델이 계속 학습을 진행하면서 점점 복잡도가 높아지는 순간이 오면, 분산이 커지면서 모델의 전체 오차가 증가된다. = 과적합

- 정규화는 분산을 낮추어 과적합을 막는 방법.
