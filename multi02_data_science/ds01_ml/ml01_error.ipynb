```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 7, 10, 13, 16])
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 10, 14, 17])
```

```python
fx01 = 3 * x + 1
```

```python
# 실제값 y - 예측값 fx01 의 차이
fx01_error = y - fx01
```

```python
fx01_error
```

```python
plt.plot(x, fx01)

for i in range(len(x)):
    hx = np.array([x[i], x[i]])
    hy = np.array([fx01[i], y[i]])
    plt.plot(hx, hy, marker="o")

plt.show()
```

```python
fx02 = np.full((x.shape), 10)
fx02
```

```python
fx02_error = y - fx02
fx02_error
```

```python
plt.plot(x, fx02)

for i in range(len(x)):
    hx = np.array([x[i] , x[i]])
    hy = np.array([fx02[i], y[i]])
    plt.plot(hx, hy, marker="o")
plt.show()
```

```python
print(np.sum(fx01_error))
print(np.sum(fx02_error))
```

```python
"""
Mean Squared Error (평균 제곱 오차)
-> ((오차들의 제곱) 들을 모두 더하자) 평균내자
"""
```

```python
def mean_squared_error(a, b, x, y):
    return ( (y - (a * x + b)) ** 2 ).mean()
```

```python
# fx01 = 3 * x + 1
mean_squared_error(3, 1, x, y)
```

```python
# fx02 = 0 * x + 10
mean_squared_error(0, 10, x, y)
```
