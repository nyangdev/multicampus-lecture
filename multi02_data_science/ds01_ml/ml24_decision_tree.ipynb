- 스무고개 처럼 동작 (기준 생성 -> 질문 -> True / False)
- 학습을 통해서 트리 형태의 기준 생성

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer.data
```

```python
cancer.target
```

```python
cancer.target_names
# 0: 악성, 1: 양성
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
df
```

```python
# 공백을 _로
df.columns = df.columns.map(lambda x: "_".join(x.split(" ")))
df
```

```python
df.info()
# 만일 숫자가 아닌 값 (NaN) / 결측치 (None) 등이 있으면 -> 숫자로 바꿔줘야 한다 or 해당 row 삭제 or ....
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df["target"] = cancer.target
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(24, 15))

for idx, column in enumerate(df.iloc[:, :-1]):
    plt.subplot(6, 5, idx+1)
    sns.histplot(data=df, x=column, hue="target")

plt.tight_layout()
plt.show()
```

```python
mal = df[df["target"] == 0]
ben = df[df["target"] == 1]
print(f"악성 : {mal.shape}")
print(f"양성 : {ben.shape}")
```

```python
# 악성 히트맵
plt.figure(figsize=(15, 15))
sns.heatmap(mal.corr(), annot=True)
```

```python
# 양성 히트맵
plt.figure(figsize=(15, 15))
sns.heatmap(ben.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# 트리의 깊이를 제한 (과적합 방지)
model = DecisionTreeClassifier(max_depth=4)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
