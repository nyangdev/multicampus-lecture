```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x : weight ( W * x + b)
x = np.arange(-10, 11, 1)
# y : mean squared error (미분할거라서 b는 생략)
y = x ** 2
```

```python
plt.plot(x, y)
plt.show()
```

```python
x_new = 10
cost = []
hy = []
# learning rate : 학습 비욜 (경사를 얼마나 내려갈건지)
learning_rate = 0.1
```

```python
for i in range(100):
    old_value = x_new
    cost.append(old_value - learning_rate * old_value)
    x_new = old_value - learning_rate * old_value
    hy.append(x_new ** 2)
```

```python
plt.plot(x, y)
plt.scatter(cost, hy)
plt.show()
```
