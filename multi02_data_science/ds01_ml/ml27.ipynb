랜덤 포레스트(random forest) : decision tree 여러개 -> bagging

```python
from sklearn.datasets import make_moons, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x, y = make_moons(n_samples=1000, noise=0.25, random_state=1)
```

```python
plt.plot(x[y==1, 0], x[y==1, 1], marker="o", linestyle="", color="red")
plt.plot(x[y==0, 0], x[y==0, 1], marker="o", linestyle="", color="blue")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators=5)
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
accuracy_score(y_test, predict)
```

```python
model.estimators_
```

```python
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
accuracy_score(y_test, predict)
```

```python
model.estimators_
```

```python
sns.barplot(x=model.feature_importances_, y=cancer.feature_names)
```
